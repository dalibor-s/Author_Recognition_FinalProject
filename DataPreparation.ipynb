```python
import numpy as np
import pandas as pd

from gutenberg.acquire import load_etext
from gutenberg.cleanup import strip_headers

```

```python
def extractParagraphs(text, paragraph_word_limit = 70):
    paragraphs = text.split('\n\n')

    filteredParagraphs = filter(lambda x: len(x.split(' ')) > paragraph_word_limit, paragraphs)

    return list(filteredParagraphs)
```

```python
# def paragraphs_number_calculator(totalParagraphs, calculated_paragraphs):

def get_paragraphs_from_books(authors_and_books = authors_and_books, paragraph_word_limit = 70, percetange_of_paragraphs = 0.1):

    dataset = []
    
    totalParagraphsPerAuthor = {}

    for author, book_ids in authors_and_books.items():
            totalParagraphsPerAuthor[author] = 0
            for book_id in book_ids:
                    book = strip_headers(load_etext(book_id)).strip()

                    paragraphs = extractParagraphs(book, paragraph_word_limit)
                    
                    totalParagraphs = len(paragraphs)
                    calculated_paragraphs_number = int(percetange_of_paragraphs * totalParagraphs)
                    
                    totalParagraphsPerAuthor[author] += totalParagraphs 
                    
                    number_of_paragraphs_taken = calculated_paragraphs_number if calculated_paragraphs_number > 10 else 10
                    print(totalParagraphs)
                    
                    selectedParagraphs = paragraphs[3 : number_of_paragraphs_taken + 3]
                    
                    for paragraph in selectedParagraphs:
                            dataset.append([author, paragraph])
                            
    print (totalParagraphsPerAuthor)
    return dataset
```

```python
dataset = pd.DataFrame(get_paragraphs_from_books())
dataset
```
